```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data.csv')
df = df.drop('Unnamed: 0', axis=1)
df = df.rename(columns= lambda x: x.replace('Object ', 'Feature-'))
df.head()
```

I want to find out which are the 4 failing equipments by looking at these 4 attributes.  
The approach I’m considering is identifying the records that have the most outlier values across the 4 columns.

```python
df['mean'] = df.mean(axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Feature-1', 'Feature-2', 'Feature-3', 'Feature-4']] = scaler.fit_transform(df[['Feature-1', 'Feature-2', 'Feature-3', 'Feature-4']])
```

```python
df["std"] = df.loc[:, df.columns[:4]].std(axis=1)
```

```python
df.sort_values('std', axis=0, ascending=False)
```

```python
import numpy as np

# Select the first 4 features as a NumPy array
features = df.iloc[:, :4].values

# Subtract Feature-1 (column 0) from all 4 features in each row
relative_features = features - features[:, [0]]

# Calculate std along axis=1 (row-wise)
df['relative_std'] = np.std(relative_features, axis=1)
```
