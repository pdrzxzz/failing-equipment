```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

"""   
Objective: Identify which equipment is likely malfunctioning by analyzing the spectral band measurements across 4 equipments.
Each row represents a spectral band; each column represents one equipment.   
"""   
   

```python
df = pd.read_csv('data.csv')
df = df.drop('Unnamed: 0', axis=1)
df = df.rename(columns= lambda x: x.replace('Object ', 'Equipment-'))
df
```

Considering that each row represents a spectral band and each column a equipment:

We need to look at the column with the most distinct values compared to the same row of other columns.

We can easily put that in numbers by calculating the mean and the standard deviation of each column.

```python
df.mean(axis=0)
```

```python
print(f'Overall average: {df.mean(axis=0).mean()}')
```

It seems that Equipment 2 and Equipment 4 have significant differences compared to the absolute mean (i.e., the mean of the column means, which is 465.9):

Equipment 2 has a mean value of 429.9, resulting in a difference of
465.9 − 429.9 = 36.0

Equipment 4 has a mean value of 503.9, resulting in a difference of
503.9 − 465.9 = 38.0

Both deviations are notable when compared to the overall average.

```python
df.std(axis=0)
```

Equipment-2 has the highest standard deviation among all, which indicates that its 17 band values vary more compared to the others.    

Final decision: **Equipment-2** is likely to be the failing one
due to its significantly higher variation across spectral bands (highest standard deviation),
and also because its average measurement is notably lower than the overall mean.
